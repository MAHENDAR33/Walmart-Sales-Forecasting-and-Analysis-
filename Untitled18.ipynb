```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/Walmart DataSet.csv")  # Update with actual file name

# Convert date column to datetime format
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")

# Checking for missing values
print(df.isnull().sum())  # Displays missing values in each column

# Filling missing values (if any)
df.fillna(df.median(), inplace=True)  # Replace with median values

# Summary statistics
print(df.describe())

# Outlier detection (Boxplot)
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['Weekly_Sales'])
plt.title("Outlier Analysis of Weekly Sales")
plt.show()
```

```python
# Correlation between Weekly Sales & Unemployment Rate
corr_unemp = df[['Weekly_Sales', 'Unemployment']].corr()
print(corr_unemp)

# Visualizing effect of Unemployment Rate on Weekly Sales
plt.figure(figsize=(10,5))
sns.scatterplot(x=df['Unemployment'], y=df['Weekly_Sales'])
plt.title("Unemployment Rate vs Weekly Sales")
plt.xlabel("Unemployment Rate")
plt.ylabel("Weekly Sales")
plt.show()
```

```python
# Extracting month & year for trend analysis
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Seasonal trend visualization
plt.figure(figsize=(12,6))
sns.lineplot(x=df['Month'], y=df['Weekly_Sales'], hue=df['Year'], palette="tab10")
plt.title("Seasonal Trends in Weekly Sales")
plt.xlabel("Month")
plt.ylabel("Weekly Sales")
plt.show()
```

```python
# Correlation between Weekly Sales & CPI
corr_cpi = df[['Weekly_Sales', 'CPI']].corr()
print(corr_cpi)

# CPI vs Weekly Sales visualization
plt.figure(figsize=(10,5))
sns.scatterplot(x=df['CPI'], y=df['Weekly_Sales'])
plt.title("Consumer Price Index vs Weekly Sales")
plt.xlabel("CPI")
plt.ylabel("Weekly Sales")
plt.show()
```

```python
# Group by store and find total sales
top_stores = df.groupby("Store")["Weekly_Sales"].sum().sort_values(ascending=False)
print("Top 5 Performing Stores:\n", top_stores.head(5))

# Visualizing top 5 stores
plt.figure(figsize=(10,5))
sns.barplot(x=top_stores.head(5).index, y=top_stores.head(5).values, palette="viridis")
plt.title("Top 5 Performing Stores")
plt.xlabel("Store Number")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Finding worst performing store
worst_store = top_stores.tail(1)
print("Worst Performing Store:\n", worst_store)

# Comparing highest & lowest performing stores
sales_gap = top_stores.iloc[0] - worst_store.iloc[0]
print("Sales Difference Between Best and Worst Store:", sales_gap)

# Visualizing top vs worst store
plt.figure(figsize=(10,5))
sns.barplot(x=['Best Store', 'Worst Store'], y=[top_stores.iloc[0], worst_store.iloc[0]], palette=["green", "red"])
plt.title("Best vs Worst Store Sales")
plt.ylabel("Total Sales")
plt.show()
```

```python
#step1:
import statsmodels.api as sm
# Group data by Store and sum Weekly Sales (for each store)
store_sales = df.groupby(["Date", "Store"])["Weekly_Sales"].sum().unstack()

# Select a specific store (e.g., Store 1) for forecasting
y = store_sales[1]  # Store 1's sales data

# Step 2: Check Stationarity Using ADF Test
from statsmodels.tsa.stattools import adfuller

def check_stationarity(series):
    result = adfuller(series.dropna())
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    if result[1] < 0.05:
        print(" Data is Stationary")
    else:
        print(" Data is NOT Stationary (Differencing Needed)")

check_stationarity(y)

#  Step 3: Apply Differencing If Needed
y_diff = y.diff().dropna()
check_stationarity(y_diff)  # Check if stationarity improves

# Step 4: Fit SARIMA Model
mod = sm.tsa.statespace.SARIMAX(y,
                                order=(4, 1, 3),  # p, d, q
                                seasonal_order=(1, 1, 0, 52),  # Seasonal component (weekly seasonality)
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results = mod.fit()

# Print model summary
print(results.summary())

# Step 5: Forecast Next 12 Weeks
forecast_steps = 12
forecast = results.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=y.index[-1], periods=forecast_steps + 1, freq="W")[1:]

# Step 6: Plot Forecast
plt.figure(figsize=(12,5))
plt.plot(y, label="Historical Sales", color='blue')
plt.plot(forecast_index, forecast.predicted_mean, label="Predicted Sales", color='red')
plt.fill_between(forecast_index,
                 forecast.conf_int().iloc[:, 0],
                 forecast.conf_int().iloc[:, 1],
                 color='red', alpha=0.2)
plt.title("Store 1 - Weekly Sales Forecast (Next 12 Weeks)")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
